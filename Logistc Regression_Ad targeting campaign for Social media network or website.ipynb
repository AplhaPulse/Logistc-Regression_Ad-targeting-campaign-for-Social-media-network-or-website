```python
#Logistic regression
S = pd.DataFrame(Social,columns = ['Gender','Age','EstimatedSalary'])
x = sm.add_constant(S)
y = Social['Purchased']
```

```python
#import libraries for training and testing
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler #Normalization
```

```python
#Normalization of variables
ss = StandardScaler()

X_scaled = ss.fit_transform(x)

X_scaled_train,X_scaled_test,y_scaled_train,y_scaled_test = train_test_split(X_scaled,y,train_size=0.33,random_state=42)
```

```python
#Training data and Testing data
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.33,random_state =42)
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
#Prediction
prediction=lr.predict(x_test)
lr.fit(X_scaled_train,y_scaled_train)
```
